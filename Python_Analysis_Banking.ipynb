```python
bins = [0,100000,300000,float('inf')]
labels = ['Low','Medium','High']
```

```python
df['Income_Band'] = pd.cut(df['Estimated_Income'], bins = bins, labels = labels, right=False)
```

```python
df['Income_Band'].value_counts().plot(kind='bar')
```

```python
categorical_cols = df[["Income_Band","BRId","GenderId","IAId","Nationality","Amount_of_Credit_Cards","Occupation","Fee_Structure","Loyalty_Classification","Properties_Owned","Risk_Weighting"]].columns
##**##Examine distribution of unique categories in categorical column**
```

```python
for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

## **Univariate Analysis**

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount_of_Credit_Cards", "Nationality", "Occupation", "Fee_Structure", "Loyalty_Classification", "Properties_Owned", "Risk_Weighting", "Income_Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

## **Bi Variate Analysis**

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount_of_Credit_Cards", "Nationality", "Occupation", "Fee_Structure", "Loyalty_Classification", "Properties_Owned", "Risk_Weighting", "Income_Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount_of_Credit_Cards", "Nationality", "Occupation", "Fee_Structure", "Loyalty_Classification", "Properties_Owned", "Risk_Weighting", "Income_Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
# Histogram plot of value counts for different Occupation

for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

## **Numerical Analysis**

```python
numerical_cols = ['Estimated_Income', 'Superannuation_Savings', 'Credit_Card_Balance', 'Bank_Loans', 'Bank_Deposits', 'Checking_Accounts', 'Saving_Accounts', 'Foreign_Currency_Account', 'Business_Lending']

# Univariate analysis and visualization
plt.figure(figsize=(20,10))
for i,col in enumerate(numerical_cols):
  plt.subplot(6,2,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.show()
```

```python
numerical_cols = ['Estimated_Income', 'Superannuation_Savings', 'Credit_Card_Balance', 'Bank_Loans', 'Bank_Deposits', 'Checking_Accounts', 'Saving_Accounts', 'Foreign_Currency_Account', 'Business_Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Insights of EDA:

- The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
